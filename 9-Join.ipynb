```python
nomes = np.array(["Alice","Bob","Charrlie"])
notas_matematicas = np.array([85,90,78])
nota_fisica = np.array([88,76,95])
```

```python
# Usando np.vstack para empilhar verticalmente os dados
dados_estruturados = np.vstack((nomes,notas_matematicas,nota_fisica))
dados_estruturados
```

```python
# Usando np.hstack para empilhar horizontalmente os dados
matriz_notas = np.column_stack((notas_matematicas,nota_fisica))
matriz_notas
```

```python
# Adicionar uma nova coluna com a mÃ©dia das notas

media_notas = np.mean(matriz_notas, axis=1)
matriz_final = np.column_stack((matriz_notas,media_notas))
matriz_final
```
