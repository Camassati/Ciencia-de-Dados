```python
import numpy as np
```

```python
array = np.array([1,2,3,4,5])
array
```

```python
soma = np.add(array, 10)
soma
```

```python
subtracao = np.subtract(array,10)
subtracao
```

```python
produto = np.multiply(array, -2)
produto
```

```python
divisao = np.divide(array, 2)
divisao
```

```python
#função de trigonometria
seno = np.sin(array)
seno
```

```python
cosseno = np.cos(array)
cosseno
```

```python
# funções estatisticas
media = np.mean(array)
media
```

```python
desvio_padrao = np.std(array)
desvio_padrao
```

```python
#7-funçoes de comparação
Comparacao_maior_que_3 = array > 3
Comparacao_maior_que_3
```

```python
#8- Combinação de Funçoes
resultado = np.sqrt(array) + np.power(array, 2)
resultado
```

```python
#9- Funções Exponenciais
exponecial = np.exp(array)
exponecial
```

```python
logaritimo = np.log(array)
logaritimo
```

```python
#11 Função de Arredondamento
arredondamento = np.around(array /2)
arredondamento
```

```python
potencia = np.power(array,3)
potencia
```
